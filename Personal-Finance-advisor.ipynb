# Univariate Analysis

```python
df.columns
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x=df['Occupation'])
plt.title('Occupation Count')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x=df['City_Tier'])
plt.title('City Tier Count')
plt.xlabel('City Tier')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.distplot(df['Income'], bins= 1000)
plt.ticklabel_format(style='plain', axis='x')
plt.ticklabel_format(style='plain', axis='y')
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Probability')
plt.show()
```

```python
df['Income'].skew()
```

The income column is positively skewed.


```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df['Income'])
plt.ticklabel_format(style='plain', axis='x')
plt.title('Boxplot of Income')
plt.show()
```

From this graph we got an idea to split the data on the basis of income into 3 major classes as lower income, middle income, higher income.

```python
Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)
IQR = Q3 - Q1
max = Q3 + 1.5 * IQR
min = Q1 - 1.5 * IQR
print( max)
```

```python
lower_income = df[df['Income'] <= 103007]
```

```python
middle_income = df[(df['Income'] > 103007) & (df['Income'] <= 250000)]
```

```python
upper_income = df[250000 < df['Income']]
```

```python
print(lower_income.shape)
print(middle_income.shape)
print(upper_income.shape)
```

```python
lower_income.head()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x= lower_income['Income'])
plt.ticklabel_format(style='plain', axis='x')
plt.title('Boxplot of Lower Income')
plt.show()
```

```python
lower_income['Income'].describe()
```

```python
Q1 = lower_income['Income'].quantile(0.25)
Q3 = lower_income['Income'].quantile(0.75)
IQR = Q3 - Q1
max = Q3 + 1.5 * IQR
print(max)
(lower_income['Income']>= max).sum()
```

There are 493 people who earns more than 226045 in Lower income class

```python
plt.figure(figsize=(8,4))
sns.boxplot(x= middle_income['Income'])
plt.ticklabel_format(style='plain', axis='x')
plt.title('Boxplot of middle Income')
plt.show()
```

```python
Q1 = middle_income['Income'].quantile(0.25)
Q3 = middle_income['Income'].quantile(0.75)
IQR = Q3 - Q1
max = Q3 + 1.5 * IQR
print(max)
(middle_income['Income']>= max).sum()
```

There are 37 people who earns more than 226045 in middle income class
