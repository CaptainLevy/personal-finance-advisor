# Data Discription

### Importing Required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/CaptainLevy/personal-finance-advisor/Shan's-work/data/raw/data.csv"
df = pd.read_csv(url)
```

###Target column

```python
df['Achieved_Goal'] = np.where(df['Income'] - df[['Rent','Loan_Repayment', 'Insurance', 'Groceries', 'Transport', 'Eating_Out','Entertainment', 'Utilities', 'Healthcare', 'Education','Miscellaneous']].sum(axis=1) >= df['Disposable_Income'], 1, 0)
```

```python
df['Achieved_Goal'].value_counts()
```

###Saving rate Column

```python
df['Saving_Rate'] = df['Disposable_Income']/df['Income']
```

```python
df['Saving_Rate'].describe()
```

```python
df.insert(19, 'Saving_Rate', df.pop('Saving_Rate'))
```

```python
df.insert( 28, 'Total_Potential_Savings', df[['Potential_Savings_Groceries', 'Saving_Rate',
       'Potential_Savings_Transport', 'Potential_Savings_Eating_Out',
       'Potential_Savings_Entertainment', 'Potential_Savings_Utilities',
       'Potential_Savings_Healthcare', 'Potential_Savings_Education',
       'Potential_Savings_Miscellaneous']].sum(axis=1))
```

```python
df.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = df.drop(['Potential_Savings_Groceries',
       'Potential_Savings_Transport', 'Potential_Savings_Eating_Out',
       'Potential_Savings_Entertainment', 'Potential_Savings_Utilities',
       'Potential_Savings_Healthcare', 'Potential_Savings_Education',
       'Potential_Savings_Miscellaneous'], axis=1, errors='ignore')
```

# Univariate Analysis

```python
df.columns
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x=df['Occupation'])
plt.title('Occupation Count')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x=df['City_Tier'])
plt.title('City Tier Count')
plt.xlabel('City Tier')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (15, 35))
for i, col in enumerate(df.columns, 1):
    plt.subplot(7, 4, i)
    sns.histplot(x = df[col], kde=True)
    plt.title(f"Histogram of {col} Data")
    plt.plot()
```

```python
df.pivot_table(index='Dependents', values='Income', columns='Occupation')
```

```python
plt.figure(figsize=(8,4))
sns.distplot(df['Income'], bins= 20)
plt.ticklabel_format(style='plain', axis='x')
plt.ticklabel_format(style='plain', axis='y')
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Probability')
plt.show()
```

```python
df['Income'].skew()
```

The income column is positively skewed.


```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df['Income'])
plt.ticklabel_format(style='plain', axis='x')
plt.title('Boxplot of Income')
plt.show()
```

From this graph we got an idea to split the data on the basis of income into 3 major classes as lower income, middle income, higher income.

```python
Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)
IQR = Q3 - Q1
max = Q3 + 1.5 * IQR
min = Q1 - 1.5 * IQR
print( max)
```

```python
lower_income = df[df['Income'] <= 103007]
```

```python
middle_income = df[(df['Income'] > 103007) & (df['Income'] <= 250000)]
```

```python
upper_income = df[250000 < df['Income']]
```

```python
print(lower_income.shape)
print(middle_income.shape)
print(upper_income.shape)
```
