# Univariate Analysis

```python
df.columns
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x=df['Occupation'])
plt.title('Occupation Count')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x=df['City_Tier'])
plt.title('City Tier Count')
plt.xlabel('City Tier')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.distplot(df['Income'], bins= 1000)
plt.ticklabel_format(style='plain', axis='x')
plt.ticklabel_format(style='plain', axis='y')
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Probability')
plt.show()
```

```python
df['Income'].skew()
```

The income column is positively skewed.


```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df['Income'])
plt.ticklabel_format(style='plain', axis='x')
plt.title('Boxplot of Income')
plt.show()
```

From this graph we got an idea to split the data on the basis of income into 3 major classes as lower income, middle income, higher income.

```python
df['Income'].max()
```

```python
lower_income = df[df['Income'] <= 1079729]
```

```python
middle_income = df[(df['Income'] > 1079729) & (df['Income'] <= 400000)]
```

```python
upper_income = df[400000 < df['Income']]
```
