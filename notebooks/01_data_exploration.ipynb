```python
plt.figure(figsize=(15, 8), dpi=100)
sns.histplot(df['Income'], bins=50, kde=True)
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(16, 10),dpi=300)
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
potential_savings_cols = ['Potential_Savings_Groceries', 'Potential_Savings_Transport', 'Potential_Savings_Eating_Out', 'Potential_Savings_Entertainment', 'Potential_Savings_Utilities',
                          'Potential_Savings_Healthcare', 'Potential_Savings_Education', 'Potential_Savings_Miscellaneous']

df['Total_Potential_Savings'] = df[potential_savings_cols].sum(axis=1)
```

```python
df.drop(columns=potential_savings_cols, inplace=True)
```

```python
df.head()
```

```python
numeric_df1 = df.select_dtypes(include=['number'])
plt.figure(figsize=(16, 10), dpi=300)
sns.heatmap(numeric_df1.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap After Adding Total Potential Savings')
plt.show()
```

```python
plt.figure(figsize=(15, 8),dpi=300)
sns.scatterplot(x='Desired_Savings', y='Total_Potential_Savings', data=df)
plt.title('Desired Savings vs Total Potential Savings')
plt.xlabel('Desired Savings')
plt.ylabel('Total Potential Savings')
plt.show()
```
